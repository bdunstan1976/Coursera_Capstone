# Results / Discussion

After loading the dataset, cleaning up the columns and rows, and visualizing the data, the following results were observed:
1) The total number of accidents where there was property damage only was over double the amount of an accident where injury occurred

2) The address type where accidents occurred was most frequently seen as being on a block.  The result showed that the amount of accidents occurring on a block was just over double the amount of accidents happening at an intersections.  The amount of accidents occuring in alleys was the least occurring.

3) The collision types that were seen more frequently were those involving parked cars.  Head on collisions were the least frequently occurring

4) The weather conditions at the time that accidents happened were most frequently seen to be clear.  This was seen to occur three times more than the next frequent cause which was seen to be when it was raining.

5) The road conditions at the time that accidents happened were most frequently seen to be dry.  The was seen to occur almost three times more than the next frequent cause which was seen to be wet.

6) The lighting conditions at the time that accidents happened were most frequently seen to be daylight.  This was seen to occur almost three times more than the next frequent cause which was see to be dark, but with the street lights on.

7) For each year from 2004 up until present, the number of accidents with property damage consistenly seemed to be double the accidents involving injury.
